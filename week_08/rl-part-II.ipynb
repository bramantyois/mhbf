## Exercise 1: Good vs. bad Döner (7 points)

Consider a hungry reinforcement learning agent standing in a street with a bad and a good Döner shop, which are at different ends of the street. The goal of this exercise is to study the effects of the discount factor on the behavior of the agent.

### 1.1 Implemente a linear Döner-world (3 points)

Write a program that implements an agent in an environment of 50 linearly arranged states (corresponding to different positions along the street). In each state, the agent can choose among 2 actions: up or down, which take it to the respective neighboring states. In every trial, it starts in a random position. If the agent walks up in the top-most state, it receives a reward of 100 (bad Döner). If it walks down in the bottom-most state, it gets a reward of 1000 (good Döner). For every step that does not lead to a reward, the agent gets a negative reward of −10 (still no Döner). If either of the Döners is received, the trial ends and the agent starts a new trial at a new random position. Initialize the Q-values with small random numbers and let the agent learn according to the SARSA algorithm:

$$
\Delta Q(s_{t},a_{t}) = \eta (r_{t}+ \gamma Q(s_{t+1},a_{t+1}) - Q(s_{t},a_{t})) \, ,
$$

with $\eta$ = 0.1. The agent should follow an $\epsilon$-greedy policy with $\epsilon$ = 0.1. At the end of each trial, store the mean of all Q-values and the number of steps of the agent.

*Hint: Terminal states in this setting require special attention. To ensure that the TD-error can go to zero in the last iteration, you can introduce two additional states at the borders of the environment whose Q-values are always 0.*

````python
def get_reward(state):
    if state == 1:
        return 1000
    elif state == 50:
        return 100
    else:
        return -10


def e_greedy(q, state, eps=0.1):
    prob = np.random.uniform(0,1)
    if prob < eps:
        return np.random.choice([0,1])
    else:
        return np.argmax(q[state, :])


def update_q(q, cur_state, next_state, cur_move, next_move, reward, eta, gamma=0.1):
    if cur_state == 51 or cur_state == 0:
        q[cur_state, cur_move] = 0
    else:
        del_q = eta*(reward + gamma*q[next_state, next_move] - q[cur_state, cur_move])
        q[cur_state, cur_move] += del_q        
    return q


def get_state(state, action):
    if action==0:
        state -= 1
    else:
        state += 1

    return state


def init_q(upper_bound=1e-6):
    ret = upper_bound * np.random.random((52,2))
    ret[0,0] = 0
    ret[0,1] = 0
    ret[-1,0] = 0
    ret[-1,1] = 0
    return ret


def run(num_trial, eta):
    q = init_q()
    
    q_means = []
    dur_counts = []

    for j in range(num_trial):
        cur_state = np.random.choice(np.arange(1,51,1))
        cur_move = np.random.choice([0,1])

        dur_count = 0

        while True:
            next_move = e_greedy(q, cur_state)
            next_state = get_state(cur_state, next_move)

            reward = get_reward(cur_state)
            q = update_q(q, cur_state, next_state, cur_move, next_move, reward, eta)

            if reward == 100 or reward == 1000:
                break
            
            cur_state = next_state
            cur_move = next_move
            
            dur_count += 1

        q_means.append(q.mean())
        dur_counts.append(dur_count)

    q_means = np.array(q_means)
    dur_counts = np.array(dur_counts)

    return q_means, dur_counts, q[1:51]

def plot_result(res, title='', max_idx=-1):
    fig, ax = plt.subplots(2, sharex=True)
    
    if max_idx==-1:
        max_idx = res[0].shape[0]

    trialx = np.arange(0, max_idx,1)

    ax[0].plot(trialx, res[0][:max_idx])
    ax[0].set_title('Mean of Q-values')
    ax[0].set_xlabel('Trial Number')
    ax[0].set_ylabel('Q-value')
    ax[0].grid()

    ax[1].plot(trialx, res[1][:max_idx])
    ax[1].set_title('Durations over Trials')
    ax[1].set_xlabel('Trial Number')
    ax[1].set_ylabel('Duration')
    ax[1].grid()
    
    fig.suptitle(title)
    fig.tight_layout()
    plt.show()

def plot_q_state(res, title=''):

    fig, ax = plt.subplots(2, sharex=True)

    states = np.arange(1, 51, 1)

    upw = res[2][:,1]
    dwr = res[2][:,0]

    ax[0].bar(states, dwr, label='downward move', alpha=0.5)
    ax[0].bar(states, upw, label='upward move', alpha=0.5)
    ax[0].legend()
    ax[0].set_title('Final Q-values')
    ax[0].set_xlabel('State Number')
    ax[0].set_ylabel('Q-value')
    ax[0].grid()

    steps_up = np.zeros(50)
    steps_dw = np.zeros(50)
    steps_up[dwr<upw] = 1
    steps_dw[dwr>upw] = -1

    ax[1].bar(states, steps_dw, alpha=0.5, label='downward move')
    ax[1].bar(states, steps_up, alpha=0.5, label='upward move')
    ax[1].set_title('Step Preference')   
    ax[1].set_xlabel('State Number')
    ax[1].set_ylabel('Step')
    ax[1].legend()
    ax[1].grid()

    fig.suptitle(title)
    fig.tight_layout()
    plt.show()
````
